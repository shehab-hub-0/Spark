```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Lecture2").getOrCreate()
```

```python
df = spark.read.text("file:///home/jovyan/work/text.txt")
```

```python
df.printSchema()
```

```python
df.show()
```

```python
from pyspark.sql.functions import length, split, col, size, trim
```

```python
df1=df.withColumnRenamed("value", "line")\
    .filter(col("line").isNotNull() & (length(trim(col("line"))) > 0))\
    .withColumn("length", length(col("line")))\
    .withColumn("words", size(split(col("line"), " ")))\
    .withColumn("first_word", split(col("line"), " ").getItem(0))
```

```python
df1.show()
```

```python
df2=df1.withColumnRenamed("line", "line1")
```

```python
df2.show()
```

```python
df.select("value").filter(col("value").isNotNull() & (length(trim(col("value"))) > 0)).show()
```
