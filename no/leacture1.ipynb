```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.getOrCreate()
sc=spark.sparkContext
```

```python
import numpy as np
```

```python
data=np.arange(1,51)
```

```python
rdd0=sc.parallelize(data)
```

```python
rdd1=rdd0.map(lambda x:x*x)
```

```python
rdd1.collect()
```

```python
rdd0.take(5)
```

```python
rdd1.cache()
```

```python
rdd1.collect()
```

```python
str(rdd1.getStorageLevel())
```

```python
rdd1.unpersist()
```

```python
str(rdd1.getStorageLevel())
```

```python
spark
```

```python
rdd1.getNumPartitions()
```

```python
partitions_data = rdd1.glom().collect()

for index, data in enumerate(partitions_data):
    print(f"Partition {index}: {data}")
```

```python
rdd2=rdd1.repartition(10)
rdd2.getNumPartitions()
```

```python
partitions_data = rdd2.glom().collect()

for index, data in enumerate(partitions_data):
    print(f"Partition {index}: {data}")
```
